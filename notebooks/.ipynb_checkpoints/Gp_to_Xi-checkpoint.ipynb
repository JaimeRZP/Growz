```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
import pandas as pd
import pymc3 as pm
import scipy as sp
import pyccl as ccl
import os
import utils
%matplotlib inline
```

```python
params = {'h': 0.6732,
        'Omega_cdm': 0.265, #0.27
        'Omega_b': 0.050,
        'n_s': 0.9665 }#

#cosmo_ccl = ccl.Cosmology(param_dict)
classy = ccl.boltzmann.classy.Class()
classy.set(params)
classy.set({ 'output':'mPk', 'P_k_max_h/Mpc': 20, 'z_max_pk': 1000})
classy.compute()
fz0 = classy.growthrate_at_z(3)
c = sp.constants.c
```

```python
def get_data(name):
    path = '/home/jaimerz/PhD/Growth/gps/'
    folder = os.path.join(path, name)
    with open(os.path.join(folder, 'trace.pkl'), 'rb') as buff:
        trace = pickle.load(buff)
    with open(os.path.join(folder, 'prediction.pkl'), 'rb') as buff:
        prediction = pickle.load(buff)
    return trace, prediction
```

```python
def get_m_s(rows):
    cols = np.transpose(rows)
    means = np.array([])
    sigmas = np.array([])
    for col in cols:
        mean = np.mean(col)
        sigma = np.std(col)
        means = np.append(means, mean)
        sigmas = np.append(sigmas, sigma)
    return means, sigmas
```

```python
f_LCDM = np.array([])
H_LCDM = np.array([])
for z in DESI_pred['z_new'].flatten():
    H = classy.Hubble(z)
    f = classy.growthrate_at_z(z)
    H_LCDM = np.append(H_LCDM, H)
    f_LCDM = np.append(f_LCDM, f)
```

```python
from scipy.integrate import odeint
from scipy.interpolate import interp1d


def wrap(f, z, Wm, q):
    dfdz = (-1/(1+z))*((3/2)*Wm(z) - f**2 - q(z)*f)
    return dfdz

def make_Wm_samples(H_samples, z_arr):
    Wm_samples = []
    for H in H_samples:
        Wm0 = 0.315
        Wm = Wm0*(H[0]/H)**2*(1+z_arr)**3
        Wm_samples.append(Wm)
    return np.array(Wm_samples)

def make_q_samples(H_samples, z_arr):
    a_arr = 1/(1+z_arr)
    x_arr = np.log(a_arr)
    
    q_samples = []
    for H in H_samples:
        comf_H = a_arr*H
        diff_comf_H = np.gradient(comf_H)/np.gradient(x_arr)
        q_midarr = 1 + diff_comf_H/comf_H
        q_samples.append(q_midarr)
    return np.array(q_samples)

def make_f_samples(H_samples, z_arr):   
    
    f_samples = []
    q_samples = make_q_samples(H_samples, z_arr)
    Wm_samples = make_Wm_samples(H_samples, z_arr)
    
    for i in np.arange(len(H_samples)):
        Wm = Wm_samples[i]
        q = q_samples[i]
        Wm_inter = interp1d(z_arr, Wm, kind='cubic', bounds_error=False, fill_value='extrapolate')
        q_inter = interp1d(z_arr, q, kind='cubic', bounds_error=False, fill_value='extrapolate')
        
        f = odeint(wrap, fz0, z_arr[::-1], args=(Wm_inter, q_inter))
        f_samples.append(f.flatten()[::-1])
        
    return Wm_samples, q_samples, np.array(f_samples)
```

```python
Wm_DESI, q_DESI, f_DESI = make_f_samples(H_DESI*1000/c, DESI_pred['z_new'].flatten())
Wm_BOSS, q_BOSS, f_BOSS = make_f_samples(H_BOSS*1000/c, BOSS_pred['z_new'].flatten())
Wm_WFIRST, q_WFIRST, f_WFIRST = make_f_samples(H_WFIRST*1000/c, WFIRST_pred['z_new'].flatten())
Wm_CC, q_CC, f_CC = make_f_samples(H_CC*1000/c, CC_pred['z_new'].flatten())
```

```python
DESI_f_m, DESI_f_s = get_m_s(f_DESI)
WFIRST_f_m, WFIRST_f_s = get_m_s(f_WFIRST)
BOSS_f_m, BOSS_f_s = get_m_s(f_BOSS)
CC_f_m, CC_f_s = get_m_s(f_CC)
```

```python
def make_Xi_samples(H_samples, z_arr, mod_f = False, mod_b = False):
    a_arr = 1/(1+z_arr)
    x_arr = np.log(a_arr)
    Xi_samples = []
    for i in np.arange(len(H_samples)):
        if mod_f == False:
            f_arr = np.array([])   
            for z in z_arr:
                f = classy.growthrate_at_z(z)
                f_arr = np.append(f_arr, f)
                
            f_diff_arr = np.gradient(f_arr)/np.gradient(x_arr)
            
        else:
            Wm_arr, q_arr, f_arr = make_f_samples(H_samples, z_arr)
            f_diff_arr = (3/2)*Wm_arr-f_arr**2-q_arr*f_arr
        
        if mod_b == False:
            Wm_arr = np.array([])
            for z in z_arr:
                Wm = classy.Om_m(z)
                Wm_arr = np.append(Wm_arr, Wm)
            
            comf_H_arr = np.array([])
            for z in z_arr:
                comf_H = (1/(1+z))*classy.Hubble(z)
                comf_H_arr = np.append(comf_H_arr, comf_H)

            comf_H_diff_arr = np.gradient(comf_H_arr)/np.gradient(x_arr)
            q_arr = 1 + (comf_H_diff_arr/comf_H_arr)
        else:
            Wm_arr = make_Wm_samples(H_samples, z_arr)
            q_arr = make_q_samples(H_samples, z_arr)
        
        Xi_arr = f_diff_arr+q_arr*f_arr+f_arr**2
        Xi_arr *= 2/(3*Wm_arr)
        
        Xi_samples.append(Xi_arr)
        
    return np.array(Xi_samples)
```

```python
DESI_Xi = make_Xi_samples(H_DESI, DESI_pred['z_new'].flatten(), mod_b = True)
BOSS_Xi = make_Xi_samples(H_BOSS, BOSS_pred['z_new'].flatten(), mod_b = True)
WFIRST_Xi = make_Xi_samples(H_WFIRST, WFIRST_pred['z_new'].flatten(), mod_b = True)
CC_Xi = make_Xi_samples(H_CC, CC_pred['z_new'].flatten(), mod_b = True)
```

```python
DESI_Xi_m, DESI_Xi_s = get_m_s(DESI_Xi)
BOSS_Xi_m, BOSS_Xi_s = get_m_s(BOSS_Xi)
WFIRST_Xi_m, WFIRST_Xi_s = get_m_s(WFIRST_Xi)
CC_Xi_m, CC_Xi_s = get_m_s(CC_Xi)
```
